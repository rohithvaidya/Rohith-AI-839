Business View
=============

Background
----------

It can be daunting for travellers to visit a new country, since they are unfamiliar with the culture and language of the new place. Currently, english widely serves as a link langauge for communication in most countries. There are gaps in communication when neither the traveller nor the native of the host country are natively english speaking. On the travellers side, he/ she may find it difficult to clearly pose his question and the host also probably understands the question in a different way. 

Problem
-------

_**What**_ _is the problem being solved?_

The application aids the a partial-english speaking person contextually convey his question better to the host, who is also a partial english speaker, in his/ her native language 

Customer
--------

_**Who**_ _it is for? Is that a user or a beneficiary? What is the problem being solved? Who it is for?_

It will be useful for people who cannot clearly communicate in english.

Value Proposition
-----------------

_Why it needs to be solved?_

1) The application helps the user communicate with someone who does not know his language and does not have a good english proficiency

2) The application overcomes your english speaking difficulties

3) When traveling, the user need not worry of finding a local to translate his english correctly to the native
